## Project's assumptions:

- Quantity of products bought in each restaurant = number on the 'votes' column
- Each aggregate_rating describes the dining experience of just 1 person
- Cost of meal for one person = average_cost_for_two / 2
- Revenue = [(average_cost_for_two / 2) x votes]

## How the analysis will play out:

It will study the impact on revenue and rating that all restaurant characteristics provided by the dataset have. 
Those characteristics, as described by the dataset, are:

- cuisines 
- city

- opentable_support

## Cuisines and their effect on revenue and rating (when studied individually):

```python
# Starting my analysis with the effect 'cuisines' have on revenue, aggreate_rating and rating_text.
# As such, first I'm separating the string inside the column 'cuisines' into different columns so that I can then work

# Now I proceed with the cuisine separation code
cuisines_dummies = df['cuisines'].str.strip("[]").str.replace("'", "").str.get_dummies(', ')

# Add the prefix 'cuisine_'
cuisines_dummies.columns = 'cuisine_' + cuisines_dummies.columns

# Concatenate with the original DataFrame
df = pd.concat([df, cuisines_dummies], axis=1)

# Drop the original "cuisines" column
df.drop('cuisines', axis=1, inplace=True)
```

```python
# Calculate the cost of a meal for one person
df['cost_per_person'] = df['average_cost_for_two'] / 2

# Calculate revenue
df['revenue'] = df['cost_per_person'] * df['votes']

df.head()
```

```python
# Create a list of dictionaries to store statistics for each cuisine
cuisine_statistics = []

cuisine_columns = df.filter(like='cuisine_').columns

# Loop through each cuisine column
for cuisine_column in cuisine_columns:
    # Extract revenue data for the current cuisine (for each column, only selects the rows in which the cuisine's present and extracts the revenue values for those rows)
    cuisine_data = df[df[cuisine_column] == 1]['revenue']

    # Calculate statistics  
    mean_revenue = cuisine_data.mean()
    median_revenue = cuisine_data.median()
    q25_revenue = cuisine_data.quantile(0.25)
    q75_revenue = cuisine_data.quantile(0.75)
    coef_variation_revenue = cuisine_data.std() / mean_revenue

    # Store statistics in a dictionary
    stats_dict = {
        'Cuisine': cuisine_column,
        'Mean Revenue': mean_revenue,
        'Median Revenue': median_revenue,
        '25th Percentile Revenue': q25_revenue,
        '75th Percentile Revenue': q75_revenue,
        'Coefficient of Variation': coef_variation_revenue
    }

    # Append the dictionary to the list
    cuisine_statistics.append(stats_dict)

    #print("\nCuisine:", stats_dict['Cuisine'])
    #print("Mean Revenue:", stats_dict['Mean Revenue'])
    #print("Median Revenue:", stats_dict['Median Revenue'])
    #print("25th Percentile Revenue:", stats_dict['25th Percentile Revenue'])
    #print("75th Percentile Revenue:", stats_dict['75th Percentile Revenue'])
    #print("Coefficient of Variation:", stats_dict['Coefficient of Variation'])

```

```python
# Sort cuisine statistics by mean revenue in descending order
sorted_cuisine_statistics = sorted(cuisine_statistics, key=lambda x: x['Mean Revenue'], reverse=True)

# Select the top 10 cuisines
top_10_cuisines = sorted_cuisine_statistics[:10]

# Extract relevant information for plotting (via for loops)
cuisines = [entry['Cuisine'] for entry in top_10_cuisines]
mean_values = [entry['Mean Revenue'] for entry in top_10_cuisines]
median_values = [entry['Median Revenue'] for entry in top_10_cuisines]
q25_values = [entry['25th Percentile Revenue'] for entry in top_10_cuisines]
q75_values = [entry['75th Percentile Revenue'] for entry in top_10_cuisines]
coef_values = [entry['Coefficient of Variation'] for entry in top_10_cuisines]

# Create a bar chart for mean revenue
fig, ax1 = plt.subplots(figsize=(12, 8))

# Bar chart for mean revenue
ax1.bar(cuisines, mean_values, color='skyblue', label='Mean Revenue')
ax1.set_xlabel('Cuisine')
ax1.set_ylabel('Mean Revenue', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_xticklabels(cuisines, rotation=90, ha='center')

# Scatter plots for other metrics
ax2 = ax1.twinx()
ax2.scatter(cuisines, median_values, color='red', s=100, marker='o', label='Median')
ax2.scatter(cuisines, q25_values, color='green', s=100, marker='o', label='25th Percentile')
ax2.scatter(cuisines, q75_values, color='purple', s=100, marker='o', label='75th Percentile')
ax2.set_ylabel('Revenue', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Line plot for coefficient of variation
ax3 = ax1.twinx()
ax3.plot(cuisines, coef_values, color='orange', marker='o', label='Coefficient of Variation')
ax3.set_ylabel('Coefficient of Variation', color='orange')
ax3.tick_params(axis='y', labelcolor='orange')
ax3.spines['right'].set_position(('outward', 60))

# Remove the y-axis on the right side
ax2.yaxis.set_visible(False)

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax2.legend(lines + lines2 + lines3, labels + labels2 + labels3, loc='upper left')

# Set title
plt.title('Top 10 Cuisines Metrics')

# Show the plot
plt.tight_layout()
plt.show()
```

## Conclusion on the Cuisine (individual) effect on Revenue:

- Tex Mex clearly is the cuisine that's correlated with the highest point of revenue, on average, per restaurant
- Moreover, Tex Mex's revenue per restaurant tends to be high even for the lesser-performing restaurants in this category (see 25th percentile)
- The latter bullet point contributes to the variability of data points inside this cluster being low
- All in all, Tex Mex is correlated with reliabily-high revenue
